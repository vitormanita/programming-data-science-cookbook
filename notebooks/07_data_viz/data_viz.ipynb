#### Heatmap

````python
df = pd.DataFrame(
    data=[
        ("1", 0.74, 0.62, 0.12),
        ("2", 0.53, -0.11, 0.99),
        ("3", -0.4, 0.2, 0.5),
        ("4", -0.88, -1.58, 0.19),
        ("5", -0.23, -0.11, 0.35),
        ("6", 0.08, 0.37, 0.28),
        ("7", -2.16, -1.21, 0.45),
        ("8", 1.47, 1.1, 0.56),
        ("9", 0.31, -0.47, 0.72),
        ("10", 0.53, -0.1, 0.52),],
    columns = ["id", "k1", "k2", "k3"]
    )
````

````python
corr = df.corr()
corr
````

````python
sns.heatmap(data=corr)
````

````python
sns.heatmap(data=corr, annot=True)
````

## Figure or Axis?
You may have noticed by now that we have several methods to build plots. Two common ones are by using `plt.plot()` or `ax.plot()`. <br>
As an advice, you can just use `plt.` for convenience, when you are just doing experiments (quick stuff) or a single plot. If you need to customize specific things, such as custom axis (not just rotation, but values themselves) or you want to build multiple plots, then the you should access the axes property, even if it's just 1 plot.

````python
data = list(range(11))
````

````python
plt.figure(figsize=(5, 5))
plt.plot(data)
plt.show()
````

````python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.plot(data)
plt.show()
````

## Multiple plots
To build multiple plots, you can use the `fig, ax = plt.subplots(nrows, ncols)` command. You just need to specify how many rows and columns of plots you want, and where each plot fits. This is easier with seaborn due to the `ax` argument

````python
df = pd.DataFrame({
    "k1": range(2001, 2011),
    "k2": [40,  45, 44,  47, 50, 46, 42, 46, 43,  38],
    "k3": [187, 170, 188, 188, 112, 158, 165, 139, 187, 146],
})
````

**With Matplotlib**

````python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax[0].plot(df["k1"], df["k2"])
ax[1].plot(df["k1"], df["k3"])
plt.show()
````

**With Seaborn**

````python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
sns.lineplot(data=df, x="k1", y="k2", ax=ax[0])
sns.lineplot(data=df, x="k1", y="k3", ax=ax[1])
plt.show()
````

To change specifics to each plot, you just need to access its position

````python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
sns.lineplot(data=df, x="k1", y="k2", color="red", ax=ax[0])
sns.lineplot(data=df, x="k1", y="k3", color="blue", ax=ax[1])

ax[0].set_title("plot1")
ax[1].set_title("plot2")
plt.suptitle("Two plots")

ax[0].tick_params(labelrotation=45)

plt.show()
````

## Exporting plots
To export your figures, use `plt.savefig("name.format")`. <br>
You can use the `dpi` (*dots per inch*) argument to increase the quality of plots when saved. A high dpi value such as 200 or more indicates a high resolution image, while a low dpi value such as 10 indicates a lower resolution image.

````python
plt.style.use("default")
plt.figure()
sns.barplot(x=["A", "B", "C"], y=[500, 300, 100])
plt.savefig("myplot.png")
````

````python
# Check that the file exists
import os
print(os.path.exists("myplot.png"))
print(os.path.getsize("myplot.png")) # Size in bytes
````

With `dpi`

````python
plt.figure()
sns.barplot(x=["A", "B", "C"], y=[500, 300, 100])
plt.savefig("myplot.png", dpi=500)
````

````python
# Check that the file exists
import os
print(os.path.exists("myplot.png"))
print(os.path.getsize("myplot.png")) # Size in bytes (Much higher now)
````
